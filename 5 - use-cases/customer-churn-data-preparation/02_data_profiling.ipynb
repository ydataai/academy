### Import needed packages

```python
import os
import json
import warnings
import pandas as pd
from pathlib import Path

from ydata.labs.datasources import DataSources
from ydata.profiling import ProfileReport

from functions.saving_functions import read_file
```

```python
#get the dataset from the reading step
dataset = read_file('dataset.pkl')
```

## Profiling the data

```python
df = dataset.to_pandas()
```

```python
profile_title = f"Original dataset: Customer Churn"

profile = ProfileReport(df=df, title=profile_title)
profile.config.html.navbar_show = False

profile.to_file("profile.html")
profile.to_file("profile.json")
```

## Create pipeline outputs

```python
import json

profile_pipeline_output = {
    'outputs' : [
        {
      'type': 'web-app',
      'storage': 'inline',
      'source': profile.to_html(),
    }
    ]
  }

with open('mlpipeline-ui-metadata.json', 'w') as metadata_file:
    json.dump(profile_pipeline_output, metadata_file)
```
