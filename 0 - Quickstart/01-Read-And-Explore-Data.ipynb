## **YData - Read and Explore Data**

In this quickstart example, we demonstrate how to read a dataset from a YData data source, and how to generate the profiling report for the data.
<br />
For this example, we use the Adult Census Income dataset, which is a collection of census data from 1994 mainly used for prediction tasks (https://archive.ics.uci.edu/ml/datasets/adult).

In order to access a YData data source we must first create through Fabric UI a connector that specifies the source of the data. Afterward, we can access the Data Catalog and create a new Dataset from the previously defined connector. The Dataset will take a few minutes to be ready. Once it is available, we can access it and check the overview, profiling, and metadata information. Next to all these tabs, we also have a button with the text "Explore in Labs" (see the image below). After clicking on it, a snippet will be shown with the quickstart code needed to access that Dataset through a data source in a Lab. We will use the code from that snippet in the following cells, and we will also present the overview and profiling information.

We first need to import the necessary YData packages.

```python
from ydata.labs import DataSources
from ydata.profiling import ProfileReport
```

Afterward, we can read the Dataset from the data source.

```python
datasource = DataSources.get(uid='{uid}', namespace='{namespace}')
dataset = datasource.dataset
```

Finally, we can also obtain the calculated Metadata to present the profile overview information.

```python
metadata = datasource.metadata
print(metadata)
```

If we want to have access to the profiling information of the Dataset, we can generate a new Profile Report with the data. After it is ready, we can either save it to a file or directly present it in the notebook.

```python
profile = ProfileReport(dataset.to_pandas(), title="Census Profiling Report")
profile.to_file("Census-Profiling-Report.html")
profile.to_notebook_iframe()
```
